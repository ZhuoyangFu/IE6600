```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('athlete_events-6.csv')
df
```

```python
df2 = df.dropna(subset=['Medal'])
df3 = df2[df2.Year >= 2016]

sports = ['Athletics','Swimming','Rowing','Football','Hockey']
df4 = df3[df3['Sport'].isin(sports)]

df4
```

```python
df5 = df4[['Age','Height','Weight']]
df5.describe()
```

```python
ax1 = sns.scatterplot(x='Height', y='Weight', data=df4)
```

```python
ax2 = sns.jointplot(x=df4['Height'], y=df4['Weight'], kind = 'hex', color = 'g')
```

```python
ax3 = sns.violinplot(x='Medal', y='Weight', hue="Sex", data=df4)
```
