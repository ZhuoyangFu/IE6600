```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame({'Name':['Ahmed','Omar','Ali','Ziad','Saiwa','Lila'],
                  'Mobile_Sales':[2540,1370,1320,2000,2100,2150],
                  'TV_Sales':[2200,1900,2150,1850,1770,2000]})
```

```python
df
```

```python
df = df.set_index("Name")
df.plot.bar(stacked = False)
plt.xlabel("Salesmen")
plt.ylabel("Sales")
```

```python
ax = df.plot.box()
```

```python
plt.pie(df['Mobile_Sales'],labels=df['Name'])
plt.legend()
plt.show()
```

```python
plt.pie(df['TV_Sales'],labels=df['Name'])
plt.legend()
plt.show()
```
