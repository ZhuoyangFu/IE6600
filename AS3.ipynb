```python
import pandas as pd
```

```python
df = pd.read_csv('athlete_events.csv')
df
```

```python
df1 = df.dropna(subset=['Medal'])
df2 = df1[df1['Year']>=2016]
df2.head()
```

```python
df2['Sport'].value_counts()
```

```python
top = ['Athletics','Swimming','Rowing','Football','Hockey']
df3 = df2[df2['Sport'].isin(top)]
df3.head()
```

```python
import matplotlib.pyplot as plt
```

```python
topsports = ['Hockey','Rowing','Athletics','Swimming','Football']
values = [99, 144, 192, 191, 106]
plt.bar(topsports, values)
plt.show()
```

```python
plt.hist(df3['Age'])
plt.ylabel('Frequency')
plt.show()
```
