```python
#Question 1
import pandas as pd

stock = pd.DataFrame()
ticker = ['fb','aapl','amzn','goog','nflx']
for i in ticker:
    df = pd.read_csv(f'{i}.csv')
    df.insert(0,'ticker',i.upper())
    stock = stock.append(df)
    
stock.head()
```

```python
stock.to_csv('stock.csv')
```

```python
#Question 2
stock = stock.assign(date=pd.to_datetime(stock['date']),volume=stock['volume'].astype(int))
stock.sort_values(['date', 'ticker'])
stock.head()
```

```python
#Question 3
stock.sort_values(by='volume',ascending=False).head(7)
```

```python
#Question 4
new_stock = stock.melt(id_vars=['ticker', 'date'],
                       value_vars=['open', 'high', 'low', 'close', 'volume'])
new_stock.head()
```
